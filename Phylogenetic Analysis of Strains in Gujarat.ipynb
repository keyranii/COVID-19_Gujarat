## Creating the Phylogenetic Tree

We are creating a phylogenetic tree to visualize the different clades present in each city of Gujarat. It can also help us understand whether similar virus strains are found geographically close to each other.

* We then created the .aln file in FASTA form using MAFFT (v7.475). We chose the FFT-NS-2 progressive method and ran it in input order.
* We created the tree using the GTRCAT model in RaxML (v10).
* Command line commands are at the end of the notebook.
